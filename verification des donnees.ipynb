```python
def num_missing(x):
    return sum(x.isnull())
def analyseCsv(df):
    print("Valeurs manquantes par colonne:")
    print(df.apply(num_missing, axis=0).head())
    print("\nValeurs dupliquees par colonne:")
    #print(df.apply(num_missing, axis=1).head()) par ligne
    #df.duplicated()
    for col in df:
        print(col,' : ', df[col].duplicated().sum())
    print('\nLes types de donnée:')
    print(df.dtypes)
    print("\nle shape du df:")
    print(df.shape)
    #print("\n donne", df[col])
```
