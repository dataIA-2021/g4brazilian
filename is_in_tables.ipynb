```python
import pandas as pd
```

```python
df_customer = pd.read_csv('olist_customers_dataset.csv')
df_geolocation_unique = pd.read_csv('olist_geolocation_dataset_unique.csv')
df_sellers = pd.read_csv('olist_sellers_dataset.csv')
```

### Liste de tous les zip codes de customer non trouvés dans geolocation

```python
my_df = df_customer[~df_customer.customer_zip_code_prefix.isin(df_geolocation_unique.geolocation_zip_code_prefix)]
print(my_df['customer_zip_code_prefix'].to_string())
print('\n total :',my_df['customer_zip_code_prefix'].count(),'valeurs non trouvées')
```

### Liste de tous les zip codes de sellers non trouvés dans geolocation

```python
my_df = df_sellers[~df_sellers.seller_zip_code_prefix.isin(df_geolocation_unique.geolocation_zip_code_prefix)]
print(my_df['seller_zip_code_prefix'].to_string())
print('\n total :',my_df['seller_zip_code_prefix'].count(),'valeurs non trouvées')
```
